## Abstract

WhatBreed is a simple tool for detecting the breed of your pet using pictures which can be found at https://whatbreed.ai. 

## Data

The data is put together from many sources across the Internet, including the [Oxford Pet Dataset](https://www.robots.ox.ac.uk/~vgg/data/pets/), [Stanford Dogs Dataset](http://vision.stanford.edu/aditya86/ImageNetDogs) as well as free image sites such as Wikimedia and Pexels, etc. Between the various datasets, this includes the [100 most popular dog breeds identified by the American Kennel Club](https://www.akc.org/expert-advice/news/most-popular-dog-breeds-of-2018/) and the [25 most popular cat breeds identified by the Cat Fanciers Association](http://cfa.org/AboutCFA/News/PressReleases/PressReleaseTopBreedsfor2018.aspx). 

## Architecture

The model was built using fastai using resnet50 convolutional neural network. No significant feature engineering was done beyond using `get_transforms()` from the fastai library.

## Application

The backend is a Dockerized web server built on Python using Starlette, which accepts up to nine pictures, forwards them to the fastai learner to predict the breeds. The backend is hosted on Render.

The frontend is a progressive web app built using VueJS and Axios that uses HTML input to open folders or access the camera to take pictures to send to the server. Being a PWA, the application is installable through Chrome, which is especially useful on smartphones.
